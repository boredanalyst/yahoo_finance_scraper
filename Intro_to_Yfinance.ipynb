```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

## Microsoft Stocks

```python
tkr = yf.Ticker("MSFT")
hist = tkr.history(interval = "1d", start = "2023-07-01")
hist = hist.reset_index()
hist["Date"] = hist["Date"].astype(str)
hist["Date"] = hist["Date"].str[0:10]
hist["Date"] = pd.to_datetime(hist["Date"])
hist
```

```python
## Plotting the Open column.

data_x = hist["Date"]
data_y = hist["Open"]

fig, ax = plt.subplots(figsize=(10,5))

ax.plot(data_x,data_y)
for ticks in ax.get_xticklabels():
  ticks.set_fontsize(8)

ax.set_ylabel("Open",fontsize=10)
ax.set_title("Microsoft's stock has been increasing since the start of July 2023", fontsize=10)

plt.show()

```
